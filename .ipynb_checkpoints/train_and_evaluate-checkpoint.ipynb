## Setup Data Fetching

```python
#overwrite chart with yfinance
import yfinance as yf
import pandas as pd
# download and save data
import os.path
import pickle
filename = "AAPL.pickle"

if os.path.isfile(filename):
    with open(filename, 'rb') as handle:
        df = pickle.load(handle)
else:
    print("File not present, will load and write")
    df = yf.download('AAPL','2018-01-01','2020-04-21')
    with open(filename, 'wb') as handle:
        pickle.dump(df, handle)
```

```python
df["datetime"] = df.index # kind of a fix bc i guess he doesnt like my index
# now i need to replace index with normal integers
df.index = range(len(df))

#also shiet doesnt like capital letters, meaning we will kick them out
#df = df.rename(columns={'Open': 'open','High':'high','Low':'low','Close':'close','Volume':'volume'})
df.columns = [name.lower() for name in df.columns]
# man dont ask but we'll need this twice for some charts later
price_history = df.copy()

# i guess we dont need datetime at all, try to leave it out
df.drop(columns=['datetime'], inplace=True)

display(df.head(3))
```

## Setup Trading Environment

```python
from tensortrade.exchanges import Exchange
from tensortrade.exchanges.services.execution.simulated import execute_order
from tensortrade.data import Stream, DataFeed, Module
from tensortrade.instruments import USD, AAPL
from tensortrade.wallets import Wallet, Portfolio
from tensortrade.environments import TradingEnvironment

#trade_data = pd.concat([
#    fetch("Coinbase", "BTC", "1h"),
#    fetch("Coinbase", "ETH", "1h")
#], axis=1)
coinbase_data = df

coinbase = Exchange("coinbase", service=execute_order)(
    Stream("USD-AAPL", list(dfjclose_preprocess)),
 #   Stream("USD-ETH", list(coinbase_data['ETH:close']))
)

#coinbase_btc = coinbase_data.loc[:, [name.startswith("BTC") for name in coinbase_data.columns]]

with Module("coinbase") as coinbase_ns:
    nodes = [Stream(name, list(coinbase_data[name])) for name in coinbase_data.columns]

feed = DataFeed([coinbase_ns])

portfolio = Portfolio(USD, [
    Wallet(coinbase, 10000 * USD),
    Wallet(coinbase, 0 * AAPL),
#    Wallet(coinbase, 5 * ETH),
])

env = TradingEnvironment(
    feed=feed,
    portfolio=portfolio,
    action_scheme='managed-risk',
    reward_scheme='risk-adjusted',
    window_size=20

)
```

```python
%%time

from tensortrade.agents import ParallelDQNAgent

def create_env():
    feed = DataFeed([coinbase_ns])

    portfolio = Portfolio(USD, [
        Wallet(coinbase, 10000 * USD),
        Wallet(coinbase, 0 * AAPL),
    ])
    
    env = TradingEnvironment(
        feed=feed,
        portfolio=portfolio,
        action_scheme='managed-risk',
        reward_scheme='risk-adjusted',
        window_size=20
    )
    
    return env

agent = ParallelDQNAgent(create_env)

scorePDQN = agent.train(n_episodes=1800,  save_path="./agents/") # 10 eps = 1 min
```

```python
agent.save("agents/justinmodel")
```
